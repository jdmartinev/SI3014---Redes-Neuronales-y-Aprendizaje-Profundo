```python
#import libraries
from numpy.lib.function_base import append
import numpy as np


# Function to define the perceptron
def perceptron(input_data, weights, bias):
    weighted_sum = np.dot(input_data, weights) + bias
    return 1 if weighted_sum > 0 else 0


# Function to train the perceptron
def train_perceptron(input_data, target_output, learning_rate, epochs):
    # Initialize weights incluiding bias
    weights = np.random.rand(len(input_data)-1)
    bias = weights[-1]
    weights = weights[0:2]

    # Training the perceptron
    for epoch in range(epochs):
       for i in range(len(input_data)):
            output = perceptron(input_data[i], weights, bias)

            # Update weights and bias
            error = target_output[i] - output
            #print(error)
            weights += learning_rate * error * input_data[i]
            bias += learning_rate * error

    weights = append(weights, bias)
    print('the final weights vector after training is:', weights)
    return weights


# Function to test the perceptrona after training
def test_perceptron(logic_gate, input_data, weights):
    print(f"Results for {logic_gate} gate:")

    # Evaluate the perceptron
    for i in range(len(input_data)):
        bias = weights[-1]
        weights_check = weights[0:2]
        output = perceptron(input_data[i], weights_check, bias)
        print(f"Input: {input_data[i]}, Output: {output}")


# Define Hyperparameters
learning_rate = 0.1
epochs = 5000

# Define training data
input = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])


# Training and testing AND1
AND1_output = np.array([1, 0, 0, 0])
weights_bias_AND1 = train_perceptron(input, AND1_output, learning_rate, epochs)
test_perceptron("AND1", input, weights_bias_AND1)

# Write here the code lines to Training and testing AND2


# Write here the code lines to Training and testing XNOR as OR[AND1,AND2]
```
